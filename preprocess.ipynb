```python
def get_dur_time(start, end):
    """
    返回间隔时间
    Args:
        start:开始时间 <datetime.datetime>
        end:结束时间 <datetime.datetime>
    Returns:
        返回间隔时间 (ms)
    """
    return (pd.to_datetime(end) - pd.to_datetime(start)).days
```
